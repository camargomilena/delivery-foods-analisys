### Operadores de Comparação

| Operador | Nome                              | Função                                                                 |
|----------|-----------------------------------|------------------------------------------------------------------------|
| >        | Maior que                         | Verifica se um valor é maior que outro                                 |
| <        | Menor que                         | Verifica se um valor é menor que outro                                 |
| ==       | Igual a                           | Verifica se um valor é igual a outro                                   |
| !=       | Diferente de                      | Verifica se um valor é diferente de outro                              |
| >=       | Maior ou igual a                  | Verifica se um valor é maior ou igual a outro                          |
| <=       | Menor ou igual a                  | Verifica se um valor é menor ou igual a outro                          |
| and      | E                                 | Retorna True se todas as condições forem verdadeiras, caso contrário retorna False |
| or       | Ou                                | Retorna True se uma das condições for verdadeiras, caso contrário retorna False |
| not      | Não                               | Inverte o resultado: se o resultado da expressão for True, o operador retorna False |

```python
(1 > 2) and (2 < 3)
```

```python
(1 > 2) or (2 < 3)
```

```python
(1 == 2) or (2 == 3) or (4 == 4)
```

### Remover colunas do dataset

```python
dataset = dataset.drop(columns=['ID'])
```

### Criar colunas no dataset

```python
def distance(lat1, long1, lat2, long2):
    dlat = lat2 - lat1 
    dlong = long2 - long1
    distance = math.sqrt(dlat**2 + dlong**2)
    return distance * 100
```

```python
dataset['Distance_km'] = dataset.apply(lambda x: distance(x['Restaurant_latitude'], x['Restaurant_longitude'], x['Delivery_location_latitude'], x['Delivery_location_longitude']), axis=1)
dataset
```

### Melhorando dados das nossas colunas

```python
dataset['Time_taken'] = dataset['Time_taken(min)'].str.replace('(min) ', '')
```

```python
dataset['Weatherconditions'] = dataset['Weatherconditions'].str.replace('conditions ', '')
```

```python
dataset = dataset.drop(columns=['Time_taken(min)'])
```

```python
dataset
```

```python
dataset.info()
```

### Encontrando valores nulos

```python
dataset.replace({"NaN": np.nan}, regex=True, inplace = True)
```

```python
dataset.isnull().sum()
```

### Completando nossos valores inválidos com a média das nossas colunas

```python
dataset['Delivery_person_Age'] = pd.to_numeric(dataset['Delivery_person_Age'], errors='coerce')
```

```python
dataset['Delivery_person_Age'] = dataset['Delivery_person_Age'].fillna(dataset['Delivery_person_Age'].mean()).astype(int)
```

```python
dataset['Delivery_person_Ratings'] = pd.to_numeric(dataset['Delivery_person_Ratings'], errors='coerce')
```

```python
dataset['Delivery_person_Ratings'] = dataset['Delivery_person_Ratings'].fillna(dataset['Delivery_person_Ratings'].mean()).astype(float)
```

```python
dataset['Time_Orderd'] = dataset['Time_Orderd'].fillna(dataset['Time_Orderd'].mode()[0])
```

```python
dataset.info()
```

### Mudar tipos de dados

##### Mudando tipo para data

```python
dataset['Order_Date'] = pd.to_datetime(dataset['Order_Date'])
```

##### Mudando tipo para booleano

```python
dataset['Festival'] = np.where(dataset['Festival'] == 'Yes ', 1, 0)
```

```python
print(dataset['Festival'].unique())
```

```python
dataset['Festival'] = dataset['Festival'].astype('bool')
```

##### Mudando tipo para string

```python
dataset['Type_of_vehicle'] = dataset['Type_of_vehicle'].astype('string')
```
